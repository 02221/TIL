## Problem of Linear Algebra
- n linear equations, n unknowns
    - **row picture**
        - Based on coordinate system(which we are used to see)
        -  the points that satisfies both the equations is the solution  
        - each rows can make plane, and the intersection of those plane is the solution(if there is a solution)
    - **columns picture**
        - by columns 
        - linear combination  
        - basis
            - In finance, this idea can reach to concept of Arrow-Debreu securities, which span financial field

```python
A = np.array([[2, -1], [-1, 2]])
b = np.array([0, 3])
```

```python
x1 = np.linspace(-1, 3, 100)
y1 = -(A[0][0] / A[0][1]) * x1 + b[0] / A[0][1]
x2 = np.linspace(-1, 3, 100)
y2 = -(A[1][0] / A[1][1]) * x2 + b[1] / A[1][1]
fig = plt.figure(figsize = (16, 8))
ans = np.dot(np.linalg.inv(A), b)
plt.scatter(x = ans[0], y = ans[1], s = 100, c = "red")
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.xlim(-1, 3)
plt.ylim(-2, 6)
plt.title("Row Picture")
plt.show()
```

```python
origin = [0, 0]
fig = plt.figure(figsize = (16, 8))
plt.arrow(*origin, *A[:,0], head_width=0.07, head_length=0.1, lw = 2, ec='blue', fc='blue',)
plt.arrow(*origin, *A[:,1], head_width=0.07, head_length=0.1, lw = 2, ec='green', fc='green',)
plt.scatter(0, 0, s = 100, c = "black")
plt.scatter(0, 3, s = 100, c = "red")
plt.xlim(-2, 3)
plt.ylim(-2, 6)
plt.title("Column Picture")
plt.show()
```

```python
origin = [0, 0]
fig = plt.figure(figsize = (16, 8))
plt.scatter(0, 0, s = 100, c = "black")
plt.scatter(0, 3, s = 100, c = "red")
plt.arrow(*A[:,1] * 2, *A[:,0], head_width=0.07, head_length=0.05, lw = 2, ec='red', fc='red',)
plt.arrow(*A[:,1],*A[:,1] , head_width=0.07, head_length=0.05, lw = 2, ec='orange', fc='orange',)
plt.arrow(*origin, *A[:,1], head_width=0.07, head_length=0.05, lw = 2, ec='green', fc='green',)
plt.arrow(*origin, *A[:,0], head_width=0.07, head_length=0.05, lw = 2, ec='blue', fc='blue',)

plt.xlim(-3, 5)
plt.ylim(-3, 5)
plt.title("Answer of Column Picture")
plt.show()
```

## Multiplying MAtrix by a vector
- $\begin{bmatrix} 2 & 5 \\ 1 & 3 \end{bmatrix} \begin{bmatrix} 1 \\ 2 \end{bmatrix}$
    - Case1
        - Linear combination
            - Ax is a combination of columns of A
        - $1 \begin{bmatrix} 2 \\ 1 \end{bmatrix} + 2 \begin{bmatrix} 5 \\ 3 \end{bmatrix} = \begin{bmatrix} 12 \\7 \end{bmatrix}$
    - Case2
        - rowwise
            - row times x(dot product)
        -  $\begin{bmatrix} 2 & 5 \end{bmatrix} \begin{bmatrix} 1 \\ 2 \end{bmatrix} = 12$
        -  $\begin{bmatrix} 1 & 3 \end{bmatrix} \begin{bmatrix} 1 \\ 2 \end{bmatrix} = 7$
