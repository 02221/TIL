# Lecture5. Random Variable and Probability Distribution

## Terminology
- Random Variable
    - A function that assigns a real numbers to each possible outcome in the sample space of an experiment
- Discrete Random Variable
    - A random variable that con take on a coutable number of values
- Continuous Random Variable
    - A random variable whose values are uncountable
- Probability Distribution
    - A function that describing the probability for all values of the random variable

## Probability Distribution
- Common Property
    - sum(f(x)) = 1
- Discrete
    - $0 \le f(x) \le 1$ 
    - $sum(f(x)) = \sum f(x)$
- Continous
    - $0 \int^a_b f(x) \le 1$
    - $sum(f(x)) = \int f(x)$


## Describing the Probability Distribution
- Expected Value
    - $E[X] = \mu = \sum xf(x)$
        - $\int x f(x)$
- Variance
    - $V[x] = \sigma^2=E[(X-\mu)^2] = E[X^2] - \mu^2$
        - = $\sum(x-\mu)^2f(x) = \sum x^2 f(x) - \mu^2$
        - = $\int(x-\mu)^2f(x) = \int x^2 f(x) - \mu^2$
- Property 
    - Mean
        - $E[g(X)] = \sum g(x)f(x)$
            - $\int g(x)f(x)$
        - $E[aX + b] = aE[X]+b$
    - Variance
        - $V[aX + b] = a^2[X]$
    - Standard Deviation
        - $S[X] = \sigma = \sqrt{V[X]}$

## Probability Distribution
- The Joint Probability Distribution
    - $f_{XY}(x,y) = P(X=x, Y=y)$ for all (x,y)
        - $0 \le f_{XY}(x,y) \le 1$
        - $\sum \sum f_{XY}(x,y) = 1$
- The Marginal Probability Distribution
    - $f_{X}(x) = P(X=x) = \sum_{\forall y} f_{X,Y}(x,y)$
    - $f_{Y}(y) = P(Y=y) = \sum_{\forall x} f_{X,Y}(x,y)$
    

## Describing the Bivariate Distribution
- Expected Values
    - $E[X] = \mu_X = \sum_{\forall x} x f_X(x) = \sum \sum x f_{X,Y}(x,y)$
    - $E[Y] = \mu_X = \sum_{\forall y} y f_Y(y) = \sum \sum y f_{X,Y}(x,y)$
- Variance
    - $V[X] = \sigma^2_X = E[(X-\mu_X)^2] = E[X^2] - \mu_X^2$
        - $\sum(x - \mu_X)^2 f_X(x) = \sum\sum (x-\mu_X)^2f_{X,Y}(x,y)$
    - $V[Y] = \sigma^2_Y = E[(Y-\mu_Y)^2] = E[Y^2] - \mu_Y^2$    
        - $\sum(y - \mu_Y)^2 f_Y(y) = \sum\sum (y-\mu_Y)^2f_{X,Y}(x,y)$
- Covariance
    - $COV[X,Y] = \sigma_{XY} = E[(X-\mu_X)(Y-\mu_Y)] = E[XY] - \mu_X\mu_Y$
        - $\sum\sum (x-\mu_X)(y-\mu_Y)f_{X,Y}(x,y)$
        - $\sum\sum x y f_{X,Y}(x,y) - \mu_X \mu_Y$     
- Coefficient of correlation
    - $Corr[X,Y] = \rho_{XY} = {COV[X,Y] \over \sqrt{V[X]}\sqrt{V[Y]}}$
        - $-1 \le \rho_{XY} \le 1$
- Sum of two discrete random variables
    - $E[X \pm Y] = E[X] \pm E[Y]$
    - $V[X \pm Y] = V[X] + V[Y]\pm 2COV[X,Y]$
    - $V[X \pm Y] = V[X] + V[Y]$ if X and Y are independent   

## Discrete Probability Distributions
- #### The Bernoulli Distribution
    - **Def**
        - classify Success or Failure
        - probability of Success for each trial is p
    - **PMF**
        - $f(x) = p if x = 1$ else $1 - p$
    - $X \sim Bernoulli[p]$
        - $E[X] = p$
        - $V[X] = p(1-p)$
        - $S[X] = \sqrt{p(1-p)}$
- #### The Binomial Distribution        
    - **Def**
        - indepenent
        - repeated n times
    - **PMF**
        - $f(x) = \begin{pmatrix}n \\ x \end{pmatrix}p^x(1-p)^{n-x}$    
    - $X \sim B[n,p]$
        - $E[X] = np$
        - $V[X] = np(1-p)$   

- #### Poisson Distribution
    - **Def**
        - Probability of a given number of events occurring in a fixed interval of time at average $\lambda$ times
    - **PMF**
        - $f(k;\lambda) = Pr(X = k) = {\lambda^k e^{-\lambda} \over k!}$
            - e : Euler's number
                - e = $lim(1 + {1 \over n})^n$
            - k! : factorial of k
    - $X \sim Poisson(\lambda)$
        - $\lambda = E(X) = Var(X)$
            - although mean and variance is same, meaning is different
    - Property
        - Additivity
            - if $X_i$ follows independent poisson distribution, and each distribution's mean is $\lambda_i$,
                - then $E(\sum X_i) = \sum \lambda_i$
    - REaltiveness with BInomial distribution
        - $B(n, p) \rightarrow Poission(\lambda = np)$
            - where np is given
                - $n \rightarrow infty$
                - $p \rightarrow 0$

- #### Normal Distribution
    - **Def**
        - Normally distributed with parameter $\mu$ and $\sigma$
    - **PDF**
        - $f(x) = {1 \over \sqrt{2\pi}\sigma}e^{-(x-\mu)^2 \over 2\sigma^2}$    
    - $X \sim N[\mu, \sigma^2]$
        - $E[X] = \mu$
        - $V[X] = \sigma^2$
    - $aX + b \sim N[a\mu + b, a^2\sigma^2]$
        - $E[X] = \mu$        
    - $\sum X_i \sim N[\sum \mu_i, \sum \sigma^2]$

- #### Chi-Square Distribution
    - **Def**
        - Let $Z ~ N[0,1]$
            - $U=Z^2$
                - then $U$ \sim chi square distribution with 1 degress of freedom $U \sim \chi^2[1]$
        - $Z_1, Z_2, \cdots, Z_k$ where $Z ~ N[0,1]$
            - $X = Z_1^2 + Z_2^2 + \cdots + Z_k^2$
                - then $X \sim \chi^2[k] $
    - **PDF**
        - $f(x) = {1 \over \gamma({k \over 2})2^{k \over 2}}x^{{k \over 2} - 1}e^{-{x \over 2}}$
    - $X \sim \chi^2[k]$
        - $E[X] = k$
        - $V[X] = 2k$
    - Property
        - if $X_1$ and $X_2$ are independent chi-square random varialbe
            - with $k_1$ and $k_2$ DoF
        - $X_1 + X_2$ is chi-square with $k_1 + k_2$ DoF

```python
xx = np.linspace(0.01, 10, 100)
dfs = np.arange(1, 5)
lss = ["-", "--", "-.", ":"]
for df, ls in zip(dfs, lss):
    rv = sp.stats.chi2(df=df)
    plt.plot(xx, rv.pdf(xx), ls=ls, label=(f"chi^2 (DoF = {df})"))
plt.xlim(0, 10.1)
plt.ylim(0, 1)

plt.ylabel("p(x)")
plt.legend()
plt.show()
```

## t-Distribution
- Def
    - if Z and X are indepent random variabe
        - $Z \sim N[0, 1]$
        - $X \sim \chi^2[k]$
    - $T = - {Z \over \sqrt{X \over k}}$
- PDF
    - $f(t) = {\Gamma ({k + 1 \over 2}) \over \Gamma({k \over 2})}{1 \over \sqrt{k\pi}}{1 \over (1 + {t^2 \over k})^{k+1 \over 2}}$

## F-distribution
- Def
    - $X = {{U \over k_1} \over {V \over k_2}}$
        - $U \sim \chi^2[k_1]$
        - $V \sim \chi^2[k_2]$
- PDF
    - $f(x) = {\Gamma({k_1 + k_2 \over 2}) \over \Gamma{k_1 \over 2}\Gamma{k_2 \over 2}}{k_1 \over k_2}^{k1_2}x^{{k_1 \over 2} - 1}(1 + {k_1 \over k_2}x)^{{1 \over 2}(k_1 + k_2)}$
- $X \sim F[k_1, k_2]$
    - $E[X] = {k_2 \over k_2 - 2}$
    - $V[X] = {2k^2_2(k_1 + k_2 - 2 \over k_1(k_2 - 2)^2}(k_2 - 4)$
- Property
    - $X \sim t[k] \rightarrow X^2 \sim F[1 ,k]$
